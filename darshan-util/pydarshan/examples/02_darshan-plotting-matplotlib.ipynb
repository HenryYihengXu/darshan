## Plotting DarshanReports using `matplotlib`(Experimental)

This notebook demonstrates how to visualize Darshan reports using matplotlib.
The included plots are still experimental for 

```python
import darshan
darshan.enable_experimental()
```

```python
# merging records
from darshan.experimental.plots.matplotlib import plot_access_histogram
from darshan.experimental.plots.matplotlib import plot_opcounts

# prepare some reports
r1 = darshan.DarshanReport("example-logs/example.darshan", read_all=True, dtype='numpy')
r2 = darshan.DarshanReport("example-logs/example2.darshan", read_all=True, dtype='numpy')
rx = r1 + r2
```

```python
# access histograms
plt = plot_access_histogram(r1, 'POSIX')
plt.show()
```

```python
plt = plot_opcounts(r1)
plt.show()
```

```python
for r in [r1, r2, rx]:
    plt = plot_opcounts(r)
    plt.show()
```
